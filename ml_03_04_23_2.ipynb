```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pylab as plt
```

```python
base = "https://raw.githubusercontent.com/lmcanavals/ml/main/data/"
```

```python
names = ["Sepal length", "Sepal width", "Petal length", "Petal width", "Species"]
data = pd.read_csv(f"{base}iris.data", header=None, names=names)
data.head()
```

```python
data.plot(kind="scatter", x="Sepal length", y="Petal length")
```

```python
sns.FacetGrid(data, hue="Species", height=10).map(plt.scatter, "Sepal length", "Petal length").add_legend()
```
